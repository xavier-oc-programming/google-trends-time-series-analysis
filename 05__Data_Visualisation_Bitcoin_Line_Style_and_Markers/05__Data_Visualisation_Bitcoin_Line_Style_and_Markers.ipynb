## **5)— Data Visualisation – Bitcoin: Line Style and Markers**

### **Introduction**

In this section, the Tesla visualisation workflow is reused and adapted for **Bitcoin**.
Since the Bitcoin price data has already been resampled to a **monthly frequency**, it can now be directly compared with **Google News search interest** for Bitcoin on the same timeline.

The emphasis here is on **styling techniques**:

* Line styles
* Markers
* Axis limits
* Visual interpretation of price vs search behaviour

---

### **Challenge Objectives**

The chart is modified to achieve the following:

* Update the chart title to
  **“Bitcoin News Search vs Resampled Price”**
* Change the left y-axis label to **“BTC Price”**
* Adjust x-axis and y-axis limits for better readability
* Apply a **dashed linestyle** to the Bitcoin price line
* Use **markers** (small circles) for search data points
* Analyse whether spikes in search interest coincide with price spikes

---

### **Key Styling Techniques**

#### **Line Styles**

Matplotlib allows line appearance to be changed using the `linestyle` argument.

Common examples:

* Solid line
* Dashed line
* Dotted line
* Dash-dot line

In this chart:

* The **Bitcoin price** is shown as a **dashed line**
* This visually separates it from the search trend

---

#### **Markers**

Markers highlight individual data points.

Examples include:

* Points
* Circles
* Triangles (up, down, left, right)

In this visualisation:

* Search interest data uses **circle markers**
* This helps emphasise that the search data is **monthly and discrete**

---

### **Using the Correct Time Index**

A key detail in this chart:

* The x-axis uses `df_btc_monthly.index`
* The monthly Bitcoin prices already store the date in the index

This ensures:

* Correct alignment with search data
* Clean and consistent time spacing

---

### **Axis Limits and Readability**

To improve clarity:

* The y-axis for price is limited (e.g. $0 to $15,000)
* The x-axis spans only the relevant date range
* Font sizes and rotation are reused from previous charts

These adjustments prevent extreme values from compressing the rest of the data.

---

### **Interpreting the Chart**

Key observations from the final chart:

* **Late 2017 / early 2018**

  * Huge spike in both Bitcoin price and search interest
  * Corresponds to peak public attention and speculation
* **Early 2019**

  * Significant price increase
  * Much smaller increase in search interest

This suggests:

* Early spikes were driven by widespread public curiosity
* Later price movements occurred after Bitcoin was already widely known

---

### **Key Takeaways**

* Line styles help distinguish different types of data
* Markers are useful for highlighting discrete observations
* Search interest and price do not always move together
* Early market hype often produces stronger search signals than later price growth
* Careful styling improves both readability and interpretability

This completes the Bitcoin line style and marker visualisation for Day 75.
