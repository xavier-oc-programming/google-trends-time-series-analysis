## **2)— Data Cleaning – Resampling Time Series Data**

### **Introduction**

Before comparing or visualising time series data, the datasets must be **clean, consistent, and aligned in time**.
This step focuses on identifying **missing values**, correcting **data types**, and **resampling** time series so that different datasets can be compared fairly.

---

### **Step 1: Checking for Missing Values**

The first task is to check whether any of the four DataFrames contain **missing (NaN) values**.

The method used:

* `.isna()` to detect missing values
* `.values.any()` to quickly verify if any missing values exist
* `.values.sum()` to count how many missing values there are

Results:

* `df_tesla` → no missing values
* `df_unemployment` → no missing values
* `df_btc_search` → no missing values
* `df_btc_price` → **contains missing values**

---

### **Identifying Missing Data in Bitcoin Prices**

For the Bitcoin price dataset:

* Missing values were found in **2 columns (same row)**
* Both `CLOSE` and `VOLUME` were missing
* The affected date was identified using a filtered `.isna()` check

This confirms that the issue is isolated and can be safely removed.

---

### **Removing Missing Values**

To clean the dataset:

* `.dropna()` is used
* The `inplace=True` argument updates the DataFrame directly

After this step:

* `df_btc_price` no longer contains missing values
* The dataset is safe to use for resampling and analysis

---

### **Step 2: Converting Strings to DateTime Objects**

Time series operations in Pandas require **datetime objects**, not strings.

Initial check:

* The `MONTH` and `DATE` columns were stored as strings (`str`)
* Checking the type of a single value is not sufficient
* The column dtype must be verified

Conversion method:

* `pd.to_datetime()` is applied to all time-related columns

Converted columns:

* `df_tesla.MONTH`
* `df_unemployment.MONTH`
* `df_btc_search.MONTH`
* `df_btc_price.DATE`

After conversion:

* Columns are stored as `datetime64[ns]`
* The data is now compatible with time-based operations like resampling

---

### **Step 3: Resampling Time Series Data**

The Bitcoin datasets use **different time frequencies**:

* Price data → **daily**
* Search data → **monthly**

To make them comparable, the daily price data must be converted to monthly data.

---

### **Resampling from Daily to Monthly**

The Pandas `.resample()` function is used with:

* A **monthly frequency** (`'M'`)
* The `DATE` column as the time reference

Chosen strategy:

* Use the **last available price of each month**
* This represents the **month-end closing price**

Alternative approach:

* Using `.mean()` would compute the average monthly price
* Month-end prices were chosen for consistency with financial analysis

---

### **Result of Resampling**

After resampling:

* Bitcoin monthly price data has **73 rows**
* This exactly matches the **73 rows** in the Bitcoin search dataset

This alignment confirms that:

* Both datasets now share the same time frequency
* They are ready to be merged and visualised together

---

### **Key Takeaways**

* Always check for **missing values** before analysis
* Time series data must use **datetime objects**
* Datasets with different frequencies require **resampling**
* Aligning time periods is essential for valid comparisons
* Pandas `.resample()` is a core tool for time series analysis

This completes the data cleaning and preparation step for Day 75.
