### **Understanding Google Trends Search Values**

Google Trends search values:

* Range from **0 to 100**
* Are **relative**, not absolute

Meaning of values:

* **100** → Peak popularity for the selected region and time period
* **50** → Half as popular as the peak
* **0** → Insufficient data for that term

Important clarification:

* Google does **not** provide actual search volumes
* Values are **scaled proxies** for popularity

Each value represents search interest relative to:

* All searches
* In the selected region
* During the selected time period
