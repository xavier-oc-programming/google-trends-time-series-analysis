## **7)— Data Visualisation – Unemployment: The Effect of New Data**

### **Introduction**

So far, the unemployment analysis showed that the **2007–2008 financial crisis** caused a massive increase in unemployment, followed by a **slow recovery of nearly 10 years**.
In this section, new data from **2020** is added to the dataset to observe how **major new events** can dramatically change the interpretation of historical trends.

---

### **Challenge: Including 2020 Data**

The task is to:

* Load the dataset **`UE Benefits Search vs UE Rate 2004–20.csv`**
* Convert the `MONTH` column to Pandas `datetime`
* Plot unemployment search interest against the official unemployment rate
* Compare the effect of **COVID-19** with the 2008 financial crisis

---

### **Key Data Preparation Steps**

* Read the CSV into a DataFrame
* Convert the `MONTH` column using `pd.to_datetime()`
* Reuse the same dual-axis plotting structure as before
* Extend the x-axis to include 2020

This ensures the new data integrates cleanly with the existing time-series workflow.

---

### **Visual Comparison: 2008 vs 2020**

Once 2020 is included, the chart changes dramatically:

* The unemployment rate spikes **far higher and faster** than during the 2008 crisis
* Search interest for **“Unemployment Benefits”** also surges sharply
* The scale of the 2020 spike **dwarfs all previous data points**

This highlights how sensitive time-series visualisations are to **new extreme values**.

---

### **Key Observations**

* The **COVID-19 shock** caused an unemployment spike that exceeded the financial crisis
* Unlike 2008, the spike occurred **very suddenly**
* Historical trends that previously looked dominant become visually compressed once 2020 is added
* Conclusions drawn without the latest data would be **misleading**

---

### **Analytical Takeaways**

* Time-series analysis must be continuously updated as new data arrives
* Extreme events can completely reshape historical narratives
* Visualisations should always be revisited when new periods are added
* Context matters: different crises produce very different economic dynamics

---

### **Key Lesson**

New data does not just extend a chart — it can **change the story entirely**.
Including 2020 demonstrates why analysts must always question conclusions drawn from incomplete timelines.
