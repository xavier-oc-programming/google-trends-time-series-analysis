## **3)— Data Visualisation – Tesla Line Charts in Matplotlib**

### **Introduction**

This section focuses on creating and progressively improving a **dual-axis line chart** to compare:

* **Tesla stock price**
* **Tesla Google search popularity**

The goal is to visualise two related time series with **different scales** on the same timeline using Matplotlib.

---

### **Challenge: Create a Basic Dual-Axis Line Chart**

The task is to plot:

* Tesla stock price on one y-axis
* Tesla search trend on a second y-axis
* Both sharing the same x-axis (time)

Key technique:

* Use `.twinx()` to create a second y-axis
* Plot each series on its own axis

This approach allows two datasets with different units to be compared visually.

---

### **Solution: Creating the Basic Chart**

Steps involved:

* Get the current axis using `plt.gca()`
* Create a second y-axis with `.twinx()`
* Set labels for each axis
* Plot each dataset separately

At this stage, the chart is functional but minimally styled.

---

### **Adding Colour for Clarity**

To improve readability:

* Assign **different colours** to each line
* Match axis label colours to their respective lines

Colour options:

* Named colours (e.g. `"skyblue"`)
* HEX colour codes (e.g. `#E6232E`)

Both axis labels and line colours are styled using keyword arguments.

---

### **Improving Chart Size and Readability**

Further enhancements include:

* Increasing figure size using `plt.figure(figsize=(14, 8))`
* Increasing resolution using `dpi`
* Making lines thicker with `linewidth`
* Increasing font sizes for labels and ticks
* Rotating x-axis labels for better readability
* Adding a clear chart title

These changes significantly improve the chart’s clarity and presentation quality.

---

### **Controlling Axis Ranges**

To focus the chart on meaningful data ranges:

* Use `.set_ylim()` to control y-axis bounds
* Use `.set_xlim()` to control the time range displayed

Axis limits can be:

* Hard-coded values
* Dynamically calculated using `.min()` and `.max()`

---

### **Explicitly Displaying the Chart**

Instead of relying on automatic notebook rendering:

* `plt.show()` is called explicitly

This ensures:

* Clean output
* Compatibility with VS Code, PyCharm, and other non-notebook environments

---

### **Fixing the Matplotlib Datetime Warning**

A Matplotlib warning may appear related to datetime conversion.

Important points:

* This is **not an error**
* It is a future-compatibility warning

To resolve it:

* Explicitly register Pandas datetime converters for Matplotlib

This ensures clean output and avoids warnings when plotting time-based data.

---

### **Key Takeaways**

* Dual-axis plots are ideal for comparing related time series with different scales
* Styling choices (colour, size, fonts) dramatically affect readability
* Explicit rendering with `plt.show()` is best practice outside interactive notebooks
* Datetime warnings should be addressed early to avoid future issues

This completes the Tesla time-series visualisation workflow.
