## **6)— Data Visualisation – Unemployment: How to Use Grids**

### **Introduction**

This section focuses on visualising **unemployment-related search behaviour** alongside the **official U.S. unemployment rate**.
The main goal is to use **grid lines** and **rolling averages** to better identify **seasonality**, **long-term trends**, and **leading vs lagging indicators**.

---

### **Challenge: Plot Unemployment Searches vs Unemployment Rate**

The task is to visualise:

* Google search interest for **“Unemployment Benefits”**
* Official **FRED unemployment rate (UNRATE)**

Using a dual-axis chart, similar to previous sections.

---

### **Chart Requirements**

The chart is modified to include:

* Title:
  **Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate**
* Left y-axis label:
  **FRED U/E Rate**
* Adjusted axis limits for clarity
* A **grey dashed grid**
* Dashed line style for the unemployment rate
* Visual inspection for **seasonality and patterns**

---

### **Using Grid Lines to Spot Patterns**

A key addition in this section is the grid:

* Grid colour: grey
* Line style: dashed

The grid helps:

* Align peaks and troughs with specific dates
* Visually compare spikes in searches with unemployment changes
* Detect repeating seasonal patterns

---

### **Insights from the Grid-Based Chart**

Several important observations become clear:

* Many spikes in searches for **“Unemployment Benefits”** occur at **year-end**, especially in **December**
* This indicates **seasonality** in job-related searches
* The **2007–2008 financial crisis** caused a massive spike in unemployment
* It took nearly **10 years (2007–2017)** for unemployment to return to pre-crisis levels

An interesting anomaly:

* A large spike in searches at the end of **2013** was **not** matched by a spike in unemployment
* This suggests external factors influencing search behaviour beyond unemployment itself

---

### **Challenge: Smoothing the Data with a Rolling Average**

Monthly search data is noisy and volatile.
To reveal underlying trends, a **rolling average** is applied.

---

### **Rolling Average Solution**

A rolling average is calculated using:

* `.rolling(window=6)`
* `.mean()`

This produces a **6-month rolling average** for:

* Search interest
* Unemployment rate

The smoothed data is then plotted using the same dual-axis structure.

---

### **Interpreting the Rolling Average Chart**

The rolling average reveals a key relationship:

* Searches for **“Unemployment Benefits”** increase **before** the unemployment rate rises
* Searches also decline **before** the unemployment rate falls

This indicates that:

* Search behaviour acts as a **leading indicator**
* The official unemployment rate is a **lagging indicator**

In other words, people begin searching for unemployment-related information **before** changes are reflected in official statistics.

---

### **Key Takeaways**

* Grid lines significantly improve time-series readability
* Seasonality becomes obvious when grids and locators are combined
* Rolling averages reduce noise and reveal underlying trends
* Google search data can act as a **leading economic indicator**
* Official unemployment statistics respond more slowly to real-world changes
