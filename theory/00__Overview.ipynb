## **0)— Day 75 Goals: what you will make by the end of the day**

### **Overview**

Today’s lesson focuses on **time series data analysis and visualisation** by combining **Google Trends** with real-world datasets such as **Bitcoin prices**, **Tesla stock prices**, and **US unemployment data**.

The goal is to understand how **search interest over time** can relate to **economic and financial indicators**, while learning how to properly **clean, resample, align, and visualise time series data** using Pandas and Matplotlib.

---

### **Key Questions We Will Explore**

* How does **search volume for “Bitcoin”** relate to the **price of Bitcoin** over time?
* How does search interest for a **hot stock like Tesla** relate to its **stock price**?
* How do searches for **“Unemployment Benefits”** vary with the **actual unemployment rate** in the United States?

These questions help illustrate how **public interest** and **real-world data** can move together — or diverge — over time.

---

### **What You Will Learn Today**

* How to make **time series data comparable** by:

  * Resampling data
  * Converting datasets to the same periodicity (e.g. daily → monthly)
* How to **fine-tune Matplotlib charts**, including:

  * Axis limits
  * Labels
  * Linestyles
  * Markers
  * Resolution (DPI)
* How to use **grids** to visually identify **seasonality** in time series data
* How to **detect and locate missing / NaN values** in a Pandas DataFrame
* How to use **Matplotlib Locators and DateFormatters** to better control the **time axis**
* How to **review and apply concepts** learned over the previous three days to **new datasets**
