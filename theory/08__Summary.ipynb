## **8)— Learning Points & Summary**

### **Overview**

This final section consolidates everything learned in **Day 75**, focusing on practical skills for **time-series analysis, cleaning, resampling, and visualisation** using Pandas and Matplotlib. The emphasis is on understanding data behavior over time and presenting it clearly and accurately.

---

### **Key Concepts Covered**

* **Using `.describe()`**

  * Quickly generate descriptive statistics
  * Useful for initial data exploration and sanity checks

* **Resampling Time Series with `.resample()`**

  * Change data periodicity (daily → monthly, etc.)
  * Essential for making datasets with different frequencies comparable

* **Working with `matplotlib.dates`**

  * Use locators and formatters to improve timeline readability
  * Control major and minor tick placement on time axes

* **Identifying Missing Values**

  * Detect NaNs using `.isna()`
  * Count missing values with `.isna().values.sum()`
  * Remove invalid rows with `.dropna()`

* **Improving Chart Resolution**

  * Adjust figure sharpness using the `dpi` parameter
  * Especially important outside interactive notebooks (VS Code, scripts)

* **Line Styling**

  * Create dashed (`'--'`) and dotted (`':'`) lines via `linestyle`
  * Improve clarity when plotting multiple series

* **Using Markers**

  * Apply markers such as `'o'`, `'^'`, etc.
  * Helpful for highlighting discrete data points

* **Fine-Tuning Chart Appearance**

  * Control axis limits, labels, line width, and colors
  * Use both named colors and HEX codes

* **Adding Grids with `.grid()`**

  * Reveal seasonality and recurring patterns
  * Makes alignment across time much clearer

---

### **Practical Takeaways**

* Time-series data often requires **cleaning and resampling** before analysis
* Visual styling is not cosmetic — it directly affects interpretability
* Search trends can act as **leading indicators** for real-world events
* Adding new data (like 2020) can **completely change conclusions**
